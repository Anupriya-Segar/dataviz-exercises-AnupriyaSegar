```python
import pandas as pd
import numpy as np

# World Happiness Report 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('/content/world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

## Task 1 — Regional Comparison Bar Chart

**What to build:** A horizontal bar chart showing the **average happiness score by region**, sorted from highest to lowest.

**Requirements:**
- Horizontal orientation (category names are long)
- Sorted by score, descending (so the happiest region is at the top)
- Zero baseline on x-axis
- At least one design choice that goes beyond the Plotly default (colour, annotation, labels, etc.)
- An insight title that answers: *which region stands out and why does it matter?*

**Hint:** Use `df.groupby('Region')['Happiness_Score'].mean()` to compute the averages.


```python
# Task 1: Regional comparison bar chart
# -------------------------------------

# Step 1: Compute average happiness score by region
region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score'))  # sort for horizontal bar

print(region_avg)

# Step 2: Build your chart
# YOUR CODE HERE

fig = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    title='Western Europe leads global happiness — Sub-Saharan Africa trails by over 2 points',
    labels={'Happiness_Score': 'Average Happiness Score (0–10)', 'Region': ''},
    color_discrete_sequence=['#2E75B6']
)

fig.update_layout(
    xaxis=dict(range=[0, 8.5]),       # Rule 1: zero baseline
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=13, color='#333333'),
    title_font=dict(family='Arial', size=14, color='#2C2C2C'),
    xaxis_gridcolor='#EEEEEE',
    yaxis_gridcolor='white',
    showlegend=False,
    margin=dict(l=10, r=120, t=60, b=40),
    height=480,
    width=900
)

fig.update_traces(
    marker_line_width=0,
    texttemplate='%{x:.2f}',
    textposition='outside',
    textfont=dict(size=11, color='#555555')
)

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=False)


# Save chart as PNG so it remains visible after downloading the notebook.
# fig.show() is intentionally omitted to avoid displaying the chart twice.

fig.write_image("task1_regional.png", scale=2)
Image("task1_regional.png")
```

## Task 2 — Bottom vs. Top: A Contrast Story

**What to build:** A bar chart that highlights the **gap between the happiest and least happy countries**, focusing on a specific insight.

**Requirements:**
- Show the **top 8 AND bottom 8 countries** together (16 bars total)
- Use **colour** to distinguish the two groups (not Plotly's default rainbow)
- Add a **visual separator or annotation** that emphasises the gap
- Insight title that tells the story of the gap

**Hint:** Use `pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])` to get both groups.

**Stretch goal:** Add a vertical reference line showing the global average.


```python
# Task 2: Top 8 vs. Bottom 8 contrast
# ------------------------------------

# Step 1: Get top and bottom countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")

# Step 2: Build your chart
# YOUR CODE HERE

pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])
fig = px.bar(
    combined,
    x='Happiness_Score',
    y='Country',
    orientation='h',
    color='Group',
    color_discrete_map={
        'Top 8':    '#2E75B6',    # blue — stands out
        'Bottom 8': '#D9534F'     # red — signals low scores
    },
    title='The happiest nations score nearly twice as high as the least happy — a 4-point gap divides two worlds',
    labels={'Happiness_Score': 'Happiness Score (0–10)', 'Country': '', 'Group': ''},
    height=580,
    width=900
)

fig.update_layout(
    xaxis=dict(range=[0, 8.5]),       # Rule 1: zero baseline — no exceptions
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=13, color='#333333'),
    title_font=dict(family='Arial', size=14, color='#2C2C2C'),
    xaxis_gridcolor='#EEEEEE',
    yaxis_gridcolor='white',
    margin=dict(l=10, r=40, t=65, b=40),
    legend=dict(orientation='h', y=1.06, x=0, title_text='', font=dict(size=12))
)

fig.update_traces(marker_line_width=0)
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=False)

# Stretch goal: global average reference line
fig.add_vline(
    x=global_avg,
    line_dash='dash',
    line_color='#888888',
    line_width=1.5,
    annotation_text=f'Global avg: {global_avg:.2f}',
    annotation_position='top right',
    annotation_font=dict(size=11, color='#555555')
)

# Save chart as PNG so it remains visible after downloading the notebook.
# fig.show() is intentionally omitted to avoid displaying the chart twice.

fig.write_image("task2_contrast.png", scale=2)
Image("task2_contrast.png")
```

## Done? Stretch Goal

If you finish both tasks with time to spare, try this:

**Task 3 (stretch):** Build a **grouped bar chart** comparing 2 sub-factors (e.g. `GDP_per_capita` and `Freedom`) across the 5 most populated regions. Use colour meaningfully and write an insight title.

Regions to include: `'Western Europe'`, `'Latin America'`, `'East Asia'`, `'Sub-Saharan Africa'`, `'South Asia'`


```python
# Stretch goal — grouped bar chart
# YOUR CODE HERE

target_regions = ['Western Europe', 'Latin America and Caribbean',
                  'East Asia', 'Sub-Saharan Africa', 'South Asia']

stretch_df = (
    df[df['Region'].isin(target_regions)]
    .groupby('Region')[['GDP', 'Freedom']]
    .mean()
    .reset_index()
    .sort_values('GDP', ascending=False)
)

stretch_long = stretch_df.melt(
    id_vars='Region',
    value_vars=['GDP', 'Freedom'],
    var_name='Factor',
    value_name='Score'
)

fig = px.bar(
    stretch_long,
    x='Region',
    y='Score',
    color='Factor',
    barmode='group',
    title='Western Europe leads on GDP; Latin America punches above its weight on freedom',
    labels={'Score': 'Average Sub-factor Score', 'Region': '', 'Factor': 'Sub-factor'},
    color_discrete_map={
        'GDP':     '#2E75B6',
        'Freedom': '#70AD47'
    },
    height=460,
    width=900
)

fig.update_layout(
    yaxis=dict(range=[0, 1.3]),       # Rule 1: zero baseline — no exceptions
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=13, color='#333333'),
    title_font=dict(family='Arial', size=14, color='#2C2C2C'),
    yaxis_gridcolor='#EEEEEE',
    xaxis_gridcolor='white',
    legend=dict(orientation='h', y=1.06, x=0, title_text=''),
    margin=dict(l=40, r=40, t=65, b=60)
)

fig.update_traces(marker_line_width=0)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=True)

# Save chart as PNG so it remains visible after downloading the notebook.
# fig.show() is intentionally omitted to avoid displaying the chart twice.

fig.write_image("task3_grouped.png", scale=2)
Image("task3_grouped.png")
```
